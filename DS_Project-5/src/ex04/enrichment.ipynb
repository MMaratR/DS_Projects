## прочитайте файл JSON, который вы сохранили в ex02

* один из столбцов имеет тип float, поэтому давайте определим его формат в pandas с помощью pd.options.display.float_format: числа с плавающей точкой должны отображаться с двумя десятичными знаками
* в модели отсутствуют значения, ничего с ними не делайте

```python
auto = pd.read_json('../data/auto.json', orient='records')
pd.options.display.float_format = '{:,.2f}'.format
auto
```

## обогатите фрейм данных, используя образец из этого фрейма данных

* создайте выборку из 200 новых наблюдений со случайным_состоянием = 21
    * в выборке не должно быть новых комбинаций номера автомобиля, марки и модели, поэтому весь набор данных будет согласован по этим параметрам
    * нет ограничений на возврат средств и штрафы, вы можете взять любое значение из этих столбцов наугад и использовать его для любого номера автомобиля

```python
new_auto = auto.sample(200, random_state=21)
new_auto['Refund'] = np.random.choice(new_auto['Refund'].unique(), size=len(new_auto))
new_auto['Fines'] = np.random.choice(new_auto['Fines'].unique(), size=len(new_auto))
new_auto
```

* объедините выборку с исходным фреймом данных в новый фрейм данных concat_rows

```python
concat_rows = pd.concat([auto, new_auto], ignore_index=True)
concat_rows
```

## обогатите фрейм данных concat_rows новым столбцом с сгенерированными данными

* создайте серию с названием «Год», используя случайные целые числа от 1980 до 2019
* используйте np.random.seed(21) перед генерацией лет
* объедините серию с фреймом данных и назовите ее fines

```python
np.random.seed(21)
Year = pd.Series(np.random.randint(1980, 2019, len(concat_rows)))
concat_rows['Year'] = Year
fines = concat_rows.copy()
fines
```

## обогатите фрейм данных данными из другого фрейма данных

* создайте новый фрейм данных с номерами автомобилей и их владельцами
    * получите самые популярные фамилии (вы можете найти файл surname.json во вложениях) в США
    * создайте новую серию с фамилиями (в них не должно быть специальных символов, таких как запятые, скобки и т. д.) из собранных вами данных. Количество должно быть равно числу уникальных номеров автомобилей в выборке (используйте random_state = 21)
    * создайте владельцев фрейма данных с 2 столбцами: номер и ФАМИЛИЯ

```python
surname = pd.read_json('../data/surname.json', orient='values')
surname.columns = surname.iloc[0]
surname.drop(0, inplace=True)

np.random.seed(21)
numbers = pd.Series(fines['CarNumber'].unique())
new_surname = pd.Series(np.random.choice(surname['NAME'], size=len(numbers)))

number_surname = pd.DataFrame()
number_surname['CarNumber'] = numbers
number_surname['Surname'] = new_surname
number_surname
```

* добавьте ещё 5 наблюдений в таблицу штрафов (придумайте свои собственные обозначения для номеров автомобилей и т. д.)

```python
new_fines = [['С735АК10RUS', 1, 1000, 'Volkswagen', 'Jetta', 2000],
             ['О504НК89RUS', 2, 2000, 'Volkswagen', 'Touareg', 2004],
             ['Е157ВХ05RUS', 1, 1000, 'Ford', 'Mondeo', 2006],
             ['Е209СУ39RUS', 2, 2000, 'Volkswagen', 'Touareg', 2002],
             ['Р978ЕА10RUS', 1, 1000, 'Ford', 'Mondeo', 2008]]
new_fines = pd.DataFrame(new_fines, columns=fines.columns)
fines = pd.concat([fines, new_fines], axis=0)
fines
```

* удалите последние 20 наблюдений из фрейма данных владельцев и добавьте 3 новых наблюдения (они не совпадают с теми, которые вы добавляете в фрейм данных штрафов)

```python
number_surname.drop(number_surname.tail(20).index, inplace=True)
number_surname
```

* объединить оба фрейма данных:
    * новый фрейм данных должен содержать только номера автомобилей, которые существуют в обоих фреймах данных

```python
frame_1 = pd.merge(left=fines, right=number_surname, on='CarNumber', how='inner')
frame_1
```

* объединить оба фрейма данных:
    * новый фрейм данных должен содержать все автомобильные номера, существующие в обоих фреймах данных

```python
frame_2 = pd.merge(left=fines, right=number_surname, on='CarNumber', how='outer')
frame_2
```

* объединить оба фрейма данных:
    * новый фрейм данных должен содержать только номера автомобилей из фрейма данных штрафов

```python
frame_3 = pd.merge(left=fines, right=number_surname, on='CarNumber', how='left')
frame_3
```

* объединить оба фрейма данных:
    * новый фрейм данных должен содержать только номера автомобилей из фрейма данных владельцев

```python
frame_4 = pd.merge(left=fines, right=number_surname, on='CarNumber', how='right')
frame_4
```

* создайте сводную таблицу из фрейма данных о штрафах, она должна выглядеть так (значения — это суммы штрафов), но со всеми годами (у вас значения могут отличаться):

```python
pd.pivot_table(fines,
               columns='Year',
               values='Fines',
               index=['Make', 'Model'],
               aggfunc={'Fines': 'sum'})
```

* сохраните фреймы данных о штрафах и владельцах в CSV-файлах без индекса

```python
fines.to_csv('../data/fines.csv', index=False)
number_surname.to_csv('../data/number_surname.csv', index=False)
```
